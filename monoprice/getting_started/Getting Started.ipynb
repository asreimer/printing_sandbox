#### **Notes**

Some note on getting the MP Select Mini up an running on a test print:

**Date:** 1 January 2021

As of this date, the printer was at firmware version "v30.50B".

Following the manual, page 7, "SETUP":
1. Move the extruder to the home position.
2. Adjust the bed height in quarter turns until a piece of paper snuggly fits between the bed and the head.
3. Use the "move" menu to lift the extruder above the bed ~3 cm.
4. Once happy with bed height, preheat the extruder and bed (I used 195 C for extruder and 60 C for bed).
5. If not already inserted, insert the filament into the printer until you cannot push anymore.
6. Wait for extruder to heat up and then use the "move" menu to increment the filament until some emerges, then wait for it to stop coming out and clean the extruder with a plastic scrape tool.
7. Print the "cat.gcode" test print and pause it after 2 layers have been laid down to inspect the layers. Watch to make sure that the first layer adheres to the bed. It seems normal for the first ~1 cm of the print to be thin or have a "tail" (I guess they didn't put a "purge wipe" in the gcode for this test print?). If you don't like this, you can manually advance the extruder with the "move" menu before starting the print and then clean the extruder with a plastic scrape tool right before it starts to print. After a few tries, I was able to have the print start fairly cleanly.
8. If the layers are too squished, lower the bed height by an eighth of a turn, reprint, pause, and examine the layers. If the layers are not straight, but wobbly, lift the bed height by an eighth of a turn, reprint, pause, and examine the layers.
9. When happy with the bed height, print the "cat.gcode". It takes 3 hours and looks like this:

#### **Installing Cura and Repetier**

##### **Cura**
This is software that parses 3D model files and "slices" them into the tool paths needed for the printer to print them.

The version of Cura that ships with the printer is quite old (https://github.com/Ultimaker/Cura/releases/tag/15.04.2). Instead, I installed the latest version from Ultimaker's website (Ultimaker Cura 4.8): https://ultimaker.com/software/ultimaker-cura

Once installed, Cura started up and walked me through:
1. Account creation.
2. Setting up the MP Select Mini as a known printer.

Then I opened the "cat.gcode" file and played around with Cura a bit. Neat.

##### **Repetier**
This is similar software to Cura and I opted not to install it at this time.

#### **Updating the firmware**

Don't do it. From everything I can find digging around online, the version of the firmware on this printer ("v30.50B") is the latest and cannot be updated using any of the "newer" firmware. 

#### **Checking Current Calibration Values**

Using a Serial monitor (I'm using the one built into the Arduino IDE), we can send G-code "M503" to see what the current calibration values are in the printer (see: https://mpselectmini.com/extruder_calibration):


    echo:Steps per unit:
    echo:  M92 X46.50 Y46.50 Z548.75 E48.50
    echo:Maximum feedrates (mm/s):
    echo:  M203 X150.00 Y150.00 Z1.50 E50.00
    echo:Maximum Acceleration (mm/s2):
    echo:  M201 X800 Y800 Z20 E10000
    echo:Accelerations: P=printing, R=retract and T=travel
    echo:  M204 P1500.00 R3000.00 T1500.00
    echo:Advanced variables: 
    S=Min feedrate (mm/s), 
    T=Min travel feedrate (mm/s), 
    B=minimum segment time (ms), 
    X=maximum XY jerk (mm/s),  
    Z=maximum Z jerk (mm/s),  
    E=maximum E jerk (mm/s)
    echo:  M205 S0.00 T0.00 B20000 X20.00 Z0.40 E5.00
    echo:Home offset (mm):
    echo:  M206 X0.00 Y0.00 Z0.00
    echo:Invert axis: M562 XYZE
    XYZABCD+--+-+-
    echo:PID settings:
    echo:  M301 P20.00 I0.02 D250.00 C100.00 L20
    echo:  M304 P10.00 I0.02 D305.40
    echo:Filament settings: Disabled
    echo:  M200 D1.75
    echo:  M200 D0
    ok N0 P15 B15


This also confirms that the V2 that I have is the "pre June 2017" model and doesn't require any firmware updates.
